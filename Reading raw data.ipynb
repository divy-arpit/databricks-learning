#Reading and viewing from csv

```python
df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/Volumes/salesdb/bronze/raw-data/source_crm/cust_info.csv")
display(df)
```

```python
logging.info("Starting to write data to bz table in overwrite mode")
df.write.mode("overwrite").saveAsTable("salesdb.bronze.source_crm_cust_info")
logging.info("data written successfully to salesdb.bronze.source_crm_cust_info")
```

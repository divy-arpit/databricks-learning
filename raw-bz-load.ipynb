##read json and load all data to bz

```python
%skip
#read json
logging.info("reading config file {}".format("config_load_data_bz.json"))
with open("config_load_data_bz.json","r") as config_file:
    config_data = json.load(config_file)
base_path = config_data['base_path']
source_crm = config_data["source_crm"]
source_erp = config_data["source_erp"]
logging.info("got all config data")
```

```python
%skip
for table_data in source_crm:
    name_of_source = table_data["name_of_source"]
    loc = table_data["loc"]
    bz_table_name = table_data["bz_table_name"]

    #start writing data to bz in ovewrite mode
    logging.info("Reading data from source {}".format(name_of_source))
    df=spark.read.format("csv").option("header","true").option("inferSchema","true").load(base_path+loc)
    logging.info("Writing data to bz table -> {}".format("salesdb.bronze."+bz_table_name))
    df.write.mode("overwrite").saveAsTable("salesdb.bronze."+bz_table_name)
    logging.info("Data written")
    logging.info("--------------------\n")
    


    
```

```python
%skip
for table_data in source_erp:
    name_of_source = table_data["name_of_source"]
    loc = table_data["loc"]
    bz_table_name = table_data["bz_table_name"]

    #start writing data to bz in ovewrite mode
    logging.info("Reading data from source {}".format(name_of_source))
    df=spark.read.format("csv").option("header","true").option("inferSchema","true").load(base_path+loc)
    logging.info("Writing data to bz table -> {}".format("salesdb.bronze."+bz_table_name))
    df.write.mode("overwrite").saveAsTable("salesdb.bronze."+bz_table_name)
    logging.info("Data written")
    logging.info("--------------------\n")
```
